# Data Visual Analysis

```python
# DB1 Visual Data Analysis

import pandas as pd
import matplotlib.pyplot as plt

# load the csv file
inputpath = './data/db1/db1-exB-gestures.csv'
df = pd.read_csv(inputpath)

MAX_SAMPLES = 20000

# extract the gesture and avg emg
emg_channels = [col for col in df.columns if "emg" in col]
df['mean_emg'] = df[emg_channels].mean(axis=1)

# plot the data (different color for each gesture)
fig, ax1 = plt.subplots(figsize=(15, 4))

# Plot average EMG signal
ax1.plot(df['mean_emg'].head(MAX_SAMPLES), color='black', linewidth=1, label='Mean EMG')
ax1.set_ylabel('Mean EMG')
ax1.set_xlabel('Time (samples)')
ax1.grid(True)

# Plot restimulus (gesture labels) on second axis
ax2 = ax1.twinx()
ax2.plot(df['gesture'].head(MAX_SAMPLES), color='red', alpha=0.5, label='Gesture')
ax2.set_ylabel('Gesture Label')

plt.title('DB1 Exercise B')
fig.tight_layout()
plt.show()
```

# DNN Classifier

```python
# DB1 Exercise B + Reconstruction of DB1

import torch
import torch.nn as nn
from dataloader import load_db, data_split_emg
from train_helpers import assess_model_nn

# Define EncoderDB1
class EncoderDB1(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(10, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 9)
        )

    def forward(self, x):
        return self.model(x)

encoder = EncoderDB1()

# Define GestureDecoder
class GestureDecoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(9, 16),
            nn.ReLU(),
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 53)
        )

    def forward(self, x):
        return self.model(x)

gesture_decoder = GestureDecoder()

# Define ReconstructionDecoder
class ReconstructionDecoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(9, 16),
            nn.ReLU(),
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 10)
        )

    def forward(self, x):
        return self.model(x)

recon_decoder = ReconstructionDecoder()

# Load DB1 (Exercise B)
df_db = load_db(db=1, exercise='B', reduce=False)
X_train, X_test, y_train, y_test = data_split_emg(df_db, test_size=0.2)

# Load model weights
encoder.load_state_dict(torch.load('./data/model_weights/encoderDB1_full_test.pth'))
gesture_decoder.load_state_dict(torch.load('./data/model_weights/gesture_decoder_full_test.pth'))
recon_decoder.load_state_dict(torch.load('./data/model_weights/recon_decoder_full_test.pth'))

# Assess the model's performance
assess_model_nn(encoder, gesture_decoder, recon_decoder, X_test, y_test)
```

# DNN Projection-Trained Classifier
