# Phase 1: Classification of EMG -> Hand Gestures with Random Forest and SVM

## Random Forest Classifier

```python
from train_helpers import assess_model
from dataloader import load_db, data_split_emg, save_model_weights, load_model_weights, has_model_weights
from sklearn.ensemble import RandomForestClassifier

# CONSTANTS
SAMPLES_PER_CLASS = 5000
MODEL_PATH = 'rfc_db1_full_test.pkl'
ESTIMATORS = 120
FORCE_RETRAIN = False

# Load DB1 (Exercise B)
df_db = load_db(db=1, exercise='B', samples_per_class=SAMPLES_PER_CLASS, reduce=False)
X_train, X_test, y_train, y_test = data_split_emg(df_db, test_size=0.2)

# Train Random Forest on DB1
if not has_model_weights(MODEL_PATH) or FORCE_RETRAIN:
    print("No model weights found. Training model...")
    model = RandomForestClassifier(n_estimators=ESTIMATORS)
    model.fit(X_train, y_train)
    save_model_weights(model, MODEL_PATH)
else:
    print("Model weights found. Loading model...")
    model = load_model_weights(MODEL_PATH)

# Assess the model's performance
assess_model(model, X_test, y_test)
```
